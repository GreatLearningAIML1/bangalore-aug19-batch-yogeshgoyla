```python
#First Quartile
Q1 = hrdata['Age'].quantile(0.25)

# Third Quartile
Q3 = hrdata['Age'].quantile(0.75)

# Computing IQR
IQR = Q3 - Q1

#Printing Values of Q1, Q3 and IQR
print('First Quartile is = ', Q1)
print('Third Quartile is = ', Q3)
print('Interquartile is = ',IQR)
```
